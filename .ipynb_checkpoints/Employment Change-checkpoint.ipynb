```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
original = pd.read_csv("employment-change-by-industry-quarterly.csv")
df = original.drop(["industry1", "industry2"], axis=1).rename(columns = {"quarter":"Year", "industry3":"Industry", "employment_change":"Change"})
df[['Year','Quarter']] = df.Year.str.split("-", expand=True)
```

```python
df = df.loc[(df.Industry=="food, beverages and tobacco") & (df.Change != "-")].reset_index(drop=True)
df
```

```python
for i in range(len(df)):
    df.Year[i] = df.Year[i][0:4]
    
```

```python
df
```

```python
df.Year = df.Year.astype(int)
df.Industry = df.Industry.astype(str)
df.Change = df.Change.astype(float)
df.info()
```

```python
df= df.loc[(df.Year >= 2016) & (df.Year < 2021)]
df = df.reset_index(drop=True)
```

```python
df.info()
```

```python
plt.figure(figsize = (15,8))
sns.set_palette("Paired", 10)
sns.barplot(data=df, x="Year", y="Change", hue="Quarter")
plt.show()
```
